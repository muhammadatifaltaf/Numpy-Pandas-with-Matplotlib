```python
df=pd.read_csv("IMDB-Movie-Data.csv",index_col="Title")
df.head()
```

```python
def rating_fun(x):
    if x>=8.1:
        return "good"
    else:
        return "bad"
```

```python
df['Rating_c']=df['Rating'].apply(rating_fun)
df.head()
```

```python
df['Rating_cc']=df['Rating'].apply(lambda x:'good'if x>=8.1 else 'bad')
df.head()
```

```python
import matplotlib.pyplot as plt
name=["Ali","Ahmed","Babar","Sidra"]
marks=[7,8,5,6]
Cgpa=[3.3,3.2,3,4]
index=np.arange(len(name))
width=0.3
plt.bar(index,marks,color='r',edgecolor='m',linewidth='2',alpha=0.7,linestyle="--",label='marks')
plt.bar(index+width,Cgpa,color='b',edgecolor='m',linewidth='2',alpha=0.7,linestyle="--",label='Cgpa')
plt.xticks (index+width,name,rotation=10)
plt.title("Students Marks")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```
